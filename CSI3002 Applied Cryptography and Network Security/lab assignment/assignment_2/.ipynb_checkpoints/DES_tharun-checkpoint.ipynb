### USER DEFINED FUNCTIONS

```python
def xor(a,b):
    if a==b:
        return "0"
    return "1"

def bulk_xor(a,b):
    result = ""
    for i,j in zip(a,b):
        result += xor(i,j)
    return result
        
def round_to_64bits(binary):
    while(len(binary)%8 !=0):
        binary = "0"+binary
    return binary

def hexa_to_bin(M):
    binary_m_arr = []
    binary_M = bin(int(M,16))[2:]
    binary_M = round_to_64bits(binary_M)
    for i in range(0,len(binary_M),8):
        binary_m_arr.append(binary_M[i:i+8])
    return binary_M, binary_m_arr

def rotate_bits_left(M,i):
    return M[i:]+M[:i]

def str_to_arr(M,n):
    arr = []
    for i in range(0,len(M),n):
        arr.append(M[i:i+n])
    return arr

def expand(R,E):
    Exp_R = ""
    for i in range(8):
        for j in range(6):
            # print(E[i][j]-1)
            Exp_R += R[E[i][j]-1]
    # print(Exp_R)
    return Exp_R

def round_to_n(binary,n):
    while(len(binary)%n !=0):
        binary = "0"+binary
    return binary
```
